```python
from sklearn.preprocessing import LabelEncoder

columns_to_check = ['Reviewer Name', 'Comments', 'Listing URL', 
                    'Listing Name', 'Host URL', 'Host Name']

correlations = {}

for col in columns_to_check:
    if df[col].dtype == 'object':
        # המרת טקסט למספרים בעזרת Label Encoding
        le = LabelEncoder()
        try:
            df_encoded = df[col].fillna("missing")  # טיפול בחסרים
            df[f'{col}_encoded'] = le.fit_transform(df_encoded)
            correlation = df[f'{col}_encoded'].corr(df['Is Superhost'])
            correlations[col] = round(correlation, 4)
        except Exception as e:
            correlations[col] = f"שגיאה: {e}"
    else:
        correlation = df[col].corr(df['Is Superhost'])
        correlations[col] = round(correlation, 4)

# הצגת התוצאות
for col, corr in correlations.items():
    print(f"{col}: קורלציה עם Is Superhost = {corr}")
```

the next step in my additional data prep is to reduce large catagorical columns: 

1.keeping the 20 biggest and most common catagories
2. replacing the rest of the catagories in the column with the word 'other'. this simplifyes the column so the model will not have to deal with hundredes of different catagories.


```python
# פונקציה שמצמצמת קטגוריות נדירות ומדפיסה סיכום
def reduce_categories(col_name, top_n=20):
    top_categories = df[col_name].value_counts().nlargest(top_n).index
    df[col_name] = df[col_name].apply(lambda x: x if x in top_categories else 'Other')

    # הצגת התוצאה
    print(f"\n✅ Reduced '{col_name}' to top {top_n} + 'Other'")
    print(f"📌 Unique categories now: {df[col_name].nunique()}")
    print("📊 Category distribution:")
    print(df[col_name].value_counts())

# הפעלת הפונקציה על העמודות שבחרת
reduce_categories('neighbourhood', top_n=20)
reduce_categories('Postal Code', top_n=20)
reduce_categories('Property Type', top_n=10)
```

* Noise and rare values that do not contribute to the model learning have been reduced.
* * reduced complexity of the data
*** laying the foundation to a clean and useful ONE-HOT ENCODING

1.Enriching

I have 2 ideas on how to enrich the model :
1.Engagement Level: Calculated as the ratio of total reviews to the number of years the host has been active, providing an insight into how engaging or popular a host or listing is over time.
2.Total Ratings Score: Created a new feature that accumulates all individual ratings into a single score. This helps in understanding the overall guest satisfaction in one metric and enhances model training by providing a composite measure of quality.

```python
df['Reviews'].dtype
df['Reviews'].head(10)
```

```python
# יצירת עמודה חדשה בלי להשתמש ב-chained assignment
df['Engagement_Level'] = df['Reviews'] / df['Host Years']

# החלפת ערכי אינסוף ב-None (שיהפכו ל-NaN ב-pandas)
df['Engagement_Level'] = df['Engagement_Level'].replace([np.inf, -np.inf], np.nan)

# מילוי ערכים חסרים באפס
df['Engagement_Level'] = df['Engagement_Level'].fillna(0)
```

```python
# הגדרת עמודות הדירוגים
ratings_columns = ['Overall Rating', 'Accuracy Rating', 'Cleanliness Rating', 'Checkin Rating', 'Communication Rating', 'Location Rating', 'Value Rating']

# טיפול בערכים חסרים בעמודות הדירוג
df[ratings_columns] = df[ratings_columns].fillna(0)

# יצירת עמודה חדשה שמצטברת את כל ערכי הדירוגים
df['Total_Ratings_Score'] = df[ratings_columns].sum(axis=1)

# הדפסת ה-DataFrame לבדיקה
print(df[['Total_Ratings_Score'] + ratings_columns].head())
```

Feature Selection: Top Important Features Only:

As part of the Feature Engineering & Feature Selection phase, Im training a new model using only the top 10 most important features.

This approach allows us to:
- Simplify the model and reduce computational cost.
- Better understand which features contribute the most to predictions.
- Test if a smaller subset of data can still yield strong predictive performance.

```python
from sklearn.ensemble import RandomForestClassifier

df_importance = df_scaled.copy()
df_importance = df_importance.dropna(subset=['Is Superhost'])

# הגדרת X ו-y
numeric_cols = df_importance.select_dtypes(include=['int64', 'float64']).columns
numeric_cols = numeric_cols.drop('Is Superhost', errors='ignore')  # לא נזרוק שגיאה אם היא לא שם

X = df_importance[numeric_cols]
y = df_importance['Is Superhost']

# אימון מודל RandomForest
importance_model = RandomForestClassifier(random_state=42)
importance_model.fit(X, y)

# חשיבות הפיצ'רים
importances = importance_model.feature_importances_
feature_importance = pd.Series(importances, index=X.columns).sort_values(ascending=False)

# תצוגה
print("🔝 Top 10 Most Important Features:")
print(feature_importance.head(10))
```
